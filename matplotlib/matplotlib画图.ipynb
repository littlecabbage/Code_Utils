# 一、画时序图和异常点

```python
real1 = pd.read_csv('~/Code_Utils/matplotlib/data/real_23.csv')
real1
```

```python
# 
def plot_data_and_score(raw_data):
    f,ax = plt.subplots(figsize=(20, 5))
    ax.plot(raw_data); 
    ax.set_title("raw data")
    
    alarm = np.array(real1[real1['is_anomaly'] == 1].index)
    ax.scatter(alarm, raw_data[alarm], color='red')
```

```python
plot_data_and_score(real1['value'])
```
